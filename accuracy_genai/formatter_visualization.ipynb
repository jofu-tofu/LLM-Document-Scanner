```python
import graphviz

# Create a new Digraph
dot = graphviz.Digraph(comment='RAG Workflow')
dot.attr('graph', layout= 'dot')

dot.attr('node', shape='box')
dot.node('start', 'START: Truth and Output Table')
dot.node('end', 'END: Updated Table')

# Circle will represent agents in my LangGraph
dot.attr('node', shape ='oval')
dot.node('A1', 'Row Coder')
dot.node('A2', 'Column Coder')
dot.node('A3', 'Instructor')
dot.node('A4', 'Supervisor')

dot.attr('node', shape='diamond')


# Arrows will represent the flow of data between agents

dot.edge('A1', 'A4')
dot.edge('A2', 'A4')
dot.edge('A3', 'A4')
dot.edge('A1', 'A2')
dot.edge('start', 'A4')
dot.edge('A4', 'end')

dot.attr('edge', style='dashed')
dot.edge('A4', 'A1')
dot.edge('A4', 'A3')




dot.render('formatter_visualization', format = 'png', cleanup = True)
dot
```

```python
import graphviz

# Create a new Digraph
dot = graphviz.Digraph(comment='RAG Workflow')

dot.attr('node', shape='box')
dot.node('start', 'START: Query and Purpose')


# Circle will represent agents in my LangGraph
dot.attr('node', shape ='oval')
dot.node('A1', 'Hypo-Output Creator')
dot.node('A2', 'Hypo-Output Evaluator')

dot.attr('node', shape='circle')
dot.node('Hypo-Bad', 'No')
dot.node('Hypo-Good', 'Yes')

dot.attr('node', shape='diamond')
dot.node('D1', 'Hypo-Doc Good?')

# Arrows will represent the flow of data between agents
dot.edge('start', 'A1', label = 'Query')
dot.edge('A1', 'A2', label = 'Hypo-Doc/Query')
dot.edge('A2', 'D1')
dot.edge('Hypo-Bad', 'A1', label = 'Critique')


dot.attr('edge', style='dashed')
dot.edge('D1', 'Hypo-Bad')
dot.edge('D1', 'Hypo-Good')



dot.render('langGraph_visualization', format = 'png', cleanup = True)
dot
```
